**1. What is the difference between a function and a method in Python?**

**Answer:** A **function** is a standalone block of code that performs a task and can be defined at module level. A **method** is a function that is attached to an object (a function defined inside a class or an object's callable attribute) and is called on that object (using dot notation).

**Example (function):**
```python
def add(a, b):
    return a + b

print(add(3, 4))  # 7
```

**Example (method):**
```python
s = "hello"
print(s.upper())  # 'HELLO'  # upper() is a string method
```


**2. Explain the concept of function arguments and parameters in Python.**

**Answer:** A **parameter** is the variable name listed in a function definition. An **argument** is the actual value passed to the function when calling it. Python supports positional arguments, keyword arguments, default parameters, variable positional arguments (`*args`) and variable keyword arguments (`**kwargs`).



**3. What are the different ways to define and call a function in Python?**

**Answer:** Common ways to define functions:
- Standard `def` function (named function).
- `lambda` (anonymous single-expression function).
- Nested functions (functions defined inside other functions).
- Methods inside classes (instance, class, static methods).

Ways to call:
- Direct call by name: `f()`
- Call a function stored in a variable: `fn = f; fn()`
- Passing functions as arguments to `map`, `filter`, etc.
- Using a lambda inline: `(lambda x: x*2)(5)`



**4. What is the purpose of the `return` statement in a Python function?**

**Answer:** `return` sends a value back to the caller and ends the function execution. If no `return` is used, the function returns `None`. Functions can return single values, multiple values (as tuples), lists, objects, etc.



**5. What are iterators in Python and how do they differ from iterables?**

**Answer:** An **iterable** is an object that can return an iterator (implements `__iter__()`), e.g., lists, tuples, strings. An **iterator** is an object that represents a stream of data and implements `__next__()` (and `__iter__()`), returning successive items until it raises `StopIteration`.



A key difference: you can call `iter()` on an iterable to get a fresh iterator; an iterator keeps internal state and is consumed as you call `next()` repeatedly.


**6. Explain the concept of generators in Python and how they are defined.**

**Answer:** A **generator** is a special kind of iterator defined using a function with `yield` statements (or a generator expression). Each `yield` pauses the function, returning a value and preserving local state â€” enabling lazy evaluation and efficient iteration over large sequences without creating the full list in memory.



**7. What are the advantages of using generators over regular functions?**

**Answer:** Advantages:
- **Memory efficiency**: produce items one-by-one instead of storing a whole list.
- **Lazy evaluation**: compute values only when requested (good for large or infinite streams).
- **Composability**: can be chained (pipeline) with other iterators/generators.
- **Simplicity**: less boilerplate to write streaming logic compared to iterator classes.



**8. What is a lambda function in Python and when is it typically used?**

**Answer:** A **lambda** creates a small anonymous function in one expression: `lambda args: expression`. It's typically used for short throwaway functions passed as arguments to higher-order functions like `map`, `filter`, `sorted`, or for simple `key=` functions.




**9. Explain the purpose and usage of the `map()` function in Python.**

**Answer:** `map(function, iterable, ...)` applies `function` to every item of the iterable(s) and returns an iterator of results (in Python 3). It's useful for element-wise transformations without writing an explicit loop.



**10. What is the difference between `map()`, `reduce()`, and `filter()` functions in Python?**

**Answer:**
- **`map(func, iterable)`**: transforms each item using `func` and returns an iterator of results.
- **`filter(func, iterable)`**: returns an iterator containing only elements for which `func(element)` is truthy.
- **`reduce(func, iterable)`** (from `functools`): cumulatively applies `func` to items of the iterable from left to right to reduce it to a single value (e.g., sum, product).

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.

````python
def sum_even(numbers):
    return sum(n for n in numbers if n % 2 == 0)

print(sum_even([1,2,3,4,5,6]))
````

2. Create a Python function that accepts a string and returns the reverse of that string.

````python
def reverse_string(s):
    return s[::-1]

print(reverse_string('hello'))
````

3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.

````python
def square_list(lst):
    return [x**2 for x in lst]

print(square_list([1,2,3,4]))
````

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

````python
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5)+1):
        if n % i == 0:
            return False
    return True

print([n for n in range(1,201) if is_prime(n)])
````

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.

````python
class Fibonacci:
    def __init__(self, terms):
        self.terms = terms
        self.a, self.b = 0, 1
        self.count = 0
    def __iter__(self):
        return self
    def __next__(self):
        if self.count >= self.terms:
            raise StopIteration
        self.count += 1
        val = self.a
        self.a, self.b = self.b, self.a + self.b
        return val

for num in Fibonacci(10):
    print(num)
````

6. Write a generator function in Python that yields the powers of 2 up to a given exponent.

````python
def powers_of_two(n):
    for i in range(n+1):
        yield 2**i

print(list(powers_of_two(5)))
````

7. Implement a generator function that reads a file line by line and yields each line as a string.

````python
import os

# Create a temporary file for demonstration
with open('sample_data.txt', 'w') as f:
    f.write("Data Line 1\n")
    f.write("Data Line 2\n")
    f.write("Data Line 3 - End\n")

def file_line_reader(filepath):
    """A generator that yields each line from a file, striping whitespace."""
    try:
        with open(filepath, 'r') as file:
            for line in file:
                yield line.strip()
    except FileNotFoundError:
        print(f"Error: File not found at {filepath}")

print("File Content (via Generator):")
for line in file_line_reader('sample_data.txt'):
    print(f"  -> {line}")

````

8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

````python
tuples = [(1,3),(2,2),(4,1)]
print(sorted(tuples, key=lambda x: x[1]))
````

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.

````python
temps_c = [0, 20, 37, 100]
temps_f = list(map(lambda c: (c*9/5)+32, temps_c))
print(temps_f)
````

10. Create a Python program that uses `filter()` to remove all the vowels from a given string.

````python
s = 'hello world'
vowels = 'aeiou'
result = ''.join(filter(lambda x: x.lower() not in vowels, s))
print(result)
````

11. Write a Python program, which returns a list with 2-tuples (order number, total cost with adjustments).

````python
orders = [[1, 2, 30], [2, 5, 10], [3, 1, 120]]
result = list(map(lambda x: (x[0], x[1]*x[2] if x[1]*x[2]>=100 else x[1]*x[2]+10), orders))
print(result)
````
